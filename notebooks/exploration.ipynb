### Nettoyage des données NULL dans le fichier transactions

```python
transactions=transactions.dropna()
transactions.isnull().sum()
```

### Modification du format de la date du dataframe transactions

```python
transactions['date'] = pd.to_datetime(transactions['date']).dt.floor('D')
```

```python
transactions.info()
transactions.head()
```

```python
produits['price'] = pd.to_numeric(produits['price'])
produits.head()
produits.info()
```

```python
clients.isnull().sum()
```

### Jointure des csv et création d'un unique csv df_lapage

```python
df_transactions_produits = transactions.merge(produits, on='id_prod', how='left')
df_transactions_produits.head()
```

```python
df_lapage = df_transactions_produits.merge(clients, on='client_id', how='left')
df_lapage.head()
```

```python
inspection(df_lapage)
```
